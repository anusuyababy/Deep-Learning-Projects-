```python
df['balcony'] = df['balcony'].fillna(2.0)
df['bath'] = df['bath'].fillna(2.0)
df['bath'] = df['bath'].astype(int)
df['balcony'] = df['balcony'].astype(int)
df.head()
```

```python
def num(x):
    try:
        float (x)
    except:
        return False
    return True
```

```python
def convert_to_num(x):
    y=x.split(' - ') 
    if (len(y)>1):
        return (float(y[0])+float(y[1]))/2
    try:
        return float(x)
    except:
        return None
```

```python
df['total_sqft'] = df.total_sqft.apply(convert_to_num)
df.head()
```

```python
df = df.dropna(subset=['total_sqft'])
df.head()
```

## TRAIN TEST SPLIT

```python
x=df.drop(['price'], axis = 1)
y=df['price']
x.shape, y.shape
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test, y_train, y_test = train_test_split(x,y, test_size =0.20, random_state = 3)
```

## ANN MODEL

```python
import keras
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import LeakyReLU, PReLU, ELU
from keras.layers import Dropout
```

```python
clf = Sequential()
clf.add(Dense(units = 200, kernel_initializer = 'he_normal', activation='relu', input_dim = 249))
clf.add(Dense(units = 400, kernel_initializer = 'he_normal', activation='relu'))
clf.add(Dense(units = 300, kernel_initializer = 'he_normal', activation='relu'))
clf.add(Dense(units = 200, kernel_initializer = 'he_normal', activation='relu'))
clf.add(Dense(units = 100, kernel_initializer = 'he_normal', activation='relu'))
clf.add(Dense(units = 1, kernel_initializer = 'glorot_normal', activation = 'linear'))
clf.compile(optimizer = 'adam', loss = 'mse', metrics=['mse','mae'])
```

```python
model = clf.fit(x_train, y_train, batch_size=10, validation_split=0.30, epochs=100)
```

```python
y_pred=clf.predict(x_test)
y_pred
```

```python
from sklearn.metrics import r2_score, mean_squared_error
r2_score(y_test, y_pred)
```
